```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_root = "../../../exam_2425/practical/micro+gas+turbine+electrical+energy+prediction/train"
files = os.listdir(data_root)
```

```python
data = pd.read_csv(os.path.join(data_root, files[0]))
data.head()
```

```python
arr = np.array(data["time"])
arr = arr[1:] - arr[:-1]
print("Time step range: ", arr.min(), arr.max())
print("Time steps: ", len(arr), (arr < 1.01).sum())
# plt.plot(arr[:5000])

plt.plot(data["input_voltage"], label="input_voltage")
plt.plot(data["el_power"]/300, label="el_power")
plt.legend()
plt.grid()
```
