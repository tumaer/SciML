```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.where((x>0)*(x<1), x, 0)

x_axis = np.linspace(-1, 2, 100)
plt.figure(figsize=(8, 4))
plt.plot(x_axis, f(x_axis))
```

```python
def pdf(x):
    return np.where((x>0.5)*(x<1.5), 1, 0)
plt.figure(figsize=(8, 4))
plt.plot(x_axis, pdf(x_axis))
```
